```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Extract relevant data
eeg_samples = data['EEGsample']
substate = data['substate']

# Identify indices for drowsiness and active states
drowsiness_indices = np.where(substate == 0)[0]
active_indices = np.where(substate == 1)[0]

# Extract the data for each state
drowsiness_data = eeg_samples[drowsiness_indices]
active_data = eeg_samples[active_indices]

# 1. Time Series Plot (Channel 1, First Sample)
plt.figure(figsize=(12, 6))
plt.plot(drowsiness_data[0, 0, :], label='Drowsiness', color='blue')
plt.plot(active_data[0, 0, :], label='Active', color='green')
plt.title("Time Series Comparison - Channel 1, First Sample")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.legend()
plt.show()

# 2. Mean Signal Comparison (Channel 1)
mean_drowsiness = np.mean(drowsiness_data[:, 0, :], axis=0)
mean_active = np.mean(active_data[:, 0, :], axis=0)

plt.figure(figsize=(12, 6))
plt.plot(mean_drowsiness, label='Drowsiness', color='blue')
plt.plot(mean_active, label='Active', color='green')
plt.title("Mean Signal Comparison - Channel 1")
plt.xlabel("Time (samples)")
plt.ylabel("Mean Amplitude")
plt.legend()
plt.show()

# 3. Power Spectral Density (PSD) Comparison (Channel 1)
f_drowsiness, Pxx_drowsiness = welch(drowsiness_data[0, 0, :], fs=1.0, nperseg=256)
f_active, Pxx_active = welch(active_data[0, 0, :], fs=1.0, nperseg=256)

plt.figure(figsize=(12, 6))
plt.semilogy(f_drowsiness, Pxx_drowsiness, label='Drowsiness', color='blue')
plt.semilogy(f_active, Pxx_active, label='Active', color='green')
plt.title("Power Spectral Density Comparison - Channel 1")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density")
plt.legend()
plt.show()

# 4. Boxplot Comparison (Channel 1, Across All Samples)
plt.figure(figsize=(8, 6))
plt.boxplot([drowsiness_data[:, 0, :].flatten(), active_data[:, 0, :].flatten()],
            labels=['Drowsiness', 'Active'])
plt.title("Boxplot Comparison - Channel 1")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, iirnotch

# Define a band-pass filter function
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data, axis=-1)
    return y

# Define a notch filter function
def apply_notch_filter(data, fs, freq=50.0, quality=30.0):
    b, a = iirnotch(freq, quality, fs)
    y = filtfilt(b, a, data, axis=-1)
    return y

# Example parameters
fs = 1000  # Sampling frequency in Hz
lowcut = 1.0  # Low cut-off frequency in Hz
highcut = 50.0  # High cut-off frequency in Hz

# Example data
t = np.arange(0, 1.0, 1/fs)
drowsiness_data = np.sin(2 * np.pi * 10 * t) + 0.5 * np.sin(2 * np.pi * 50 * t)  # Example data with 50 Hz noise
active_data = np.sin(2 * np.pi * 20 * t) + 0.5 * np.sin(2 * np.pi * 50 * t)  # Example data with 50 Hz noise

# Apply bandpass filter to drowsiness and active data
drowsiness_data_filtered = bandpass_filter(drowsiness_data, lowcut, highcut, fs)
active_data_filtered = bandpass_filter(active_data, lowcut, highcut, fs)

# Apply notch filter to remove 50 Hz power line noise
drowsiness_data_filtered = apply_notch_filter(drowsiness_data_filtered, fs)
active_data_filtered = apply_notch_filter(active_data_filtered, fs)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(t, drowsiness_data, label='Original Drowsiness Data')
plt.title('Original Drowsiness Data')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 2)
plt.plot(t, drowsiness_data_filtered, label='Filtered Drowsiness Data')
plt.title('Filtered Drowsiness Data')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 3)
plt.plot(t, active_data, label='Original Active Data')
plt.title('Original Active Data')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 4)
plt.plot(t, active_data_filtered, label='Filtered Active Data')
plt.title('Filtered Active Data')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, iirnotch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Define functions (bandpass_filter, apply_notch_filter) as in previous example

# Example feature extraction function (modify according to your needs)
def extract_features(data):
    # Example features: mean, variance, and peak-to-peak amplitude
    return [
        np.mean(data),
        np.var(data),
        np.ptp(data)
    ]

# Example data and labels
fs = 1000  # Sampling frequency in Hz
n_samples = 1000
drowsiness_data = np.random.randn(n_samples)
active_data = np.random.randn(n_samples)

# Apply preprocessing
drowsiness_data_filtered = apply_notch_filter(bandpass_filter(drowsiness_data, 1.0, 50.0, fs), fs)
active_data_filtered = apply_notch_filter(bandpass_filter(active_data, 1.0, 50.0, fs), fs)

# Feature extraction
def extract_features_from_data(data):
    return np.array([extract_features(data)])

# Extract features for each dataset
X_drowsiness = extract_features_from_data(drowsiness_data_filtered)
X_active = extract_features_from_data(active_data_filtered)

# Create labels
y_drowsiness = np.zeros(X_drowsiness.shape[0])  # 0 for drowsiness
y_active = np.ones(X_active.shape[0])  # 1 for active

# Combine datasets
X = np.vstack((X_drowsiness, X_active))
y = np.concatenate((y_drowsiness, y_active))

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred, target_names=['Drowsiness', 'Active']))

# Example prediction on new data
new_data = np.random.randn(n_samples)  # Replace with actual new data
new_data_filtered = apply_notch_filter(bandpass_filter(new_data, 1.0, 50.0, fs), fs)
new_features = extract_features_from_data(new_data_filtered)
new_features_scaled = scaler.transform(new_features)
prediction = clf.predict(new_features_scaled)
print("Prediction for new data:", "Drowsiness" if prediction[0] == 0 else "Active")
```

```python
import numpy as np
from scipy.signal import butter, filtfilt, iirnotch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Example functions for preprocessing and feature extraction
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y

def apply_notch_filter(data, fs, freq=50.0, quality=30.0):
    b, a = iirnotch(freq, quality, fs)
    y = filtfilt(b, a, data)
    return y

def extract_features(data):
    # Implement your feature extraction here
    return [
        np.mean(data),
        np.var(data),
        np.ptp(data),
        np.max(data) - np.min(data),
        np.median(data)
    ]

# Load or generate your EEG data and labels
# Example placeholders:
eeg_data = np.random.randn(1000, 100)  # Replace with actual EEG data
labels = np.random.randint(2, size=1000)  # Replace with actual labels

# Preprocessing
fs = 1000  # Example sampling frequency
filtered_data = np.array([apply_notch_filter(bandpass_filter(d, 1.0, 50.0, fs), fs) for d in eeg_data])

# Feature extraction
features = np.array([extract_features(d) for d in filtered_data])

# Split data
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Evaluate model
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))

# Real-time prediction example
new_eeg_data = np.random.randn(100)  # Replace with new EEG data
new_eeg_data_filtered = apply_notch_filter(bandpass_filter(new_eeg_data, 1.0, 50.0, fs), fs)
new_features = np.array(extract_features(new_eeg_data_filtered)).reshape(1, -1)
new_features_scaled = scaler.transform(new_features)
prediction = clf.predict(new_features_scaled)
print("Prediction for new data:", "Drowsy" if prediction[0] == 1 else "Alert")
```
